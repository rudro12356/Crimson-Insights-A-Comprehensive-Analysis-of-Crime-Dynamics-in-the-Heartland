```python
import pandas as pd
import numpy as np
# Load the dataset
file_path = 'C:/Users/m588h354/Downloads/crime_data_total_modelling (1).csv'
data = pd.read_csv(file_path)

# Display the first few rows of the dataframe to understand its structure
data
```

```python
from sklearn.preprocessing import MinMaxScaler

# Correlation Analysis
selected_indicators = ['Male_Officers', 'Female_Officers', 'Per_capita_income',  
                       'Percent_poverty', 'Number_of_schools', 'GDP', 'College', 
                       'Community_college', 'Total_employment', 'Country_inflation', 
                       'Percent_house_unit_vacant','Minimal_wage_per_hour','White_race', 
                       'Black_race', 'Hispanic_ethnicity', 'Native_American_race', 'Asian_race', 
                       'Multiracial', 'Native_Hawaian_race', 'Unemployment_persons_avg']

# Normalizing all features in the dataset using MinMaxScaler
scaler = MinMaxScaler()

# Selecting all features for normalization
features_to_normalize = data.columns.drop('Year')   # Excluding 'Year' from normalization

# Applying normalization
data_normalized = data.copy()
data_normalized[features_to_normalize] = scaler.fit_transform(data[features_to_normalize])

# Displaying the first few rows of the normalized data
data_normalized
```

```python
# Re-importing pandas and reloading the data as the code execution state was reset
import pandas as pd
import matplotlib.pyplot as plt
data_normalized[selected_indicators] = scaler.fit_transform(data[selected_indicators])

# Trend Visualization
plt.figure(figsize=(20, 15))

# Plotting each socio-economic feature
for i, feature in enumerate(selected_indicators, 1):
    plt.subplot(5, 4, i)
    plt.plot(data_normalized['Year'], data_normalized[feature], label=feature, marker='o')
    plt.title(feature)
    plt.xlabel('Year')
    plt.ylabel('Normalized Value')
    plt.tight_layout()

plt.show()
```

```python
from sklearn.ensemble import RandomForestRegressor
from joblib import dump, load

plt.figure(figsize=(30, 25))
crime_indicators = ['Murder', 'Rape', 'Robbery', 
                    'Assault', 'Burglary', 'Theft', 
                    'Motor_Vehicle_Theft', 'Arson']
# Plotting each socio-economic feature
for i, feature in enumerate(crime_indicators, 1):
    plt.subplot(5, 4, i)
    # Preparing the dataset for modeling
    X = data_normalized[selected_indicators]
    y = data_normalized[feature]

    # Splitting the dataset into training and test sets
    # Since we are dealing with time series data, the split should be based on time. 
    # The latest data will be used for testing and the rest for training.
    train_data = data_normalized[data_normalized['Year'] < data_normalized['Year'].max()]
    test_data = data_normalized[data_normalized['Year'] == data_normalized['Year'].max()]

    X_train = train_data[selected_indicators]
    y_train = train_data[feature]
    X_test = test_data[selected_indicators]
    y_test = test_data[feature]
    # Training a Random Forest Regressor
    model = RandomForestRegressor(n_estimators=100, random_state=0)
    model.fit(X_train, y_train)
    
    dump(model, feature+'_model.joblib')
    # Predicting 'Murder' for the entire dataset (including training and test sets)
    data_normalized['Predicted_'+feature] = model.predict(data_normalized[selected_indicators])

    # Plotting the actual and predicted trends for 'Murder'
    plt.plot(data_normalized['Year'], data_normalized[feature], label='Actual Rate '+feature, marker='o')
    plt.plot(data_normalized['Year'], data_normalized['Predicted_'+feature], label='Predicted Rate '+feature, marker='x')
    plt.title('Actual vs Predicted '+feature+ ' Rate Trend')
    plt.xlabel('Year')
    plt.ylabel('Normalized'+feature+ 'Rate')
    plt.legend()

plt.show()
```

```python

# Estimating future values for socio-economic features for the next five years (2023-2027)
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression

# Function to estimate future values using Polynomial Regression
def estimate_future_values_poly(data, feature, degree, start_year, end_year):
    """
    Estimate future values of a feature using Polynomial Regression.
    """
    years = data['Year'].values
    values = data[feature].values

    # Transforming data for Polynomial Regression
    poly = PolynomialFeatures(degree=degree)
    X_poly = poly.fit_transform(years.reshape(-1, 1))

    # Fitting the model
    poly_model = LinearRegression()
    poly_model.fit(X_poly, values)

    # Predicting for future years
    future_years = np.arange(start_year, end_year + 1)
    future_years_poly = poly.fit_transform(future_years.reshape(-1, 1))
    future_values = poly_model.predict(future_years_poly)

    return future_values

def estimate_future_values_linear(data, feature, start_year, end_year):
    """
    Estimate future values of a feature using a linear trend.
    """
    # Extracting the feature values and corresponding years
    years = data['Year'].values
    values = data[feature].values

    # Fitting a linear regression model
    linear_model = LinearRegression()
    linear_model.fit(years.reshape(-1, 1), values)

    # Predicting for future years
    future_years = np.arange(start_year, end_year + 1)
    future_values = linear_model.predict(future_years.reshape(-1, 1))

    return future_values

# Years to predict
future_years = np.arange(2023, 2030)

# Dictionary to hold future values of socio-economic features
future_values_linear = {}
future_values_poly = {}

# Estimating future values for each feature
for feature in selected_indicators:
    future_values_linear[feature] = estimate_future_values_linear(data_normalized, feature, future_years[0], future_years[-1])
    future_values_poly[feature] = estimate_future_values_poly(data_normalized, feature, 2, future_years[0], future_years[-1])

# Converting to DataFrame
future_values_linear_df = pd.DataFrame(future_values_linear, index=future_years)
future_values_poly_df = pd.DataFrame(future_values_poly, index=future_years)

# Predicting Murder rate for the next five years using the estimated future values of socio-economic features

# Dictionary to hold murder predictions for future years
future_murder_predictions_linear = {}
future_murder_predictions_poly = {}
model_all = load('Murder_model.joblib')
# Making predictions for each future year - LINEAR
for year in future_values_linear_df.index:
    future_data_linear = future_values_linear_df.loc[year].values.reshape(1, -1)
    predicted_murder_linear = model_all.predict(future_data_linear)[0]
    future_murder_predictions_linear[year] = predicted_murder_linear

# Making predictions for each future year - POLY
for year in future_values_poly_df.index:
    future_data_poly = future_values_poly_df.loc[year].values.reshape(1, -1)
    predicted_murder_poly = model_all.predict(future_data_poly)[0]
    future_murder_predictions_poly[year] = predicted_murder_poly


# Feature Importance Analysis from the trained model
feature_importances = pd.Series(model_all.feature_importances_, index=X_train.columns).sort_values(ascending=False)

# Plotting the predictions and feature importances
plt.figure(figsize=(20, 7))

# Plotting future murder predictions
plt.subplot(1, 3, 1)
plt.plot(list(future_murder_predictions_linear.keys()), list(future_murder_predictions_linear.values()), marker='o')
plt.title('Linear Predicted Murder Rate for 2023-2027')
plt.xlabel('Year')
plt.ylabel('Normalized Murder Rate')

# Plotting future murder predictions
plt.subplot(1, 3, 2)
plt.plot(list(future_murder_predictions_poly.keys()), list(future_murder_predictions_poly.values()), marker='o')
plt.title('Poly Predicted Murder Rate for 2023-2027')
plt.xlabel('Year')
plt.ylabel('Normalized Murder Rate')

# Plotting feature importances
plt.subplot(1, 3, 3)
feature_importances.plot(kind='bar')
plt.title('Feature Importances in Murder Prediction Model')
plt.xlabel('Features')
plt.ylabel('Relative Importance')

plt.tight_layout()
plt.show()

# Resetting the index to make 'Year' a column in the future values dataframes
future_values_linear_df = future_values_linear_df.reset_index().rename(columns={'index': 'Year'})

# # Merging the future values dataframes with the normalized data
combined_linear = pd.concat([data_normalized, future_values_linear_df])
# # The 'Year' column is retained in the merged dataframes
```
